```python
df.isnull().sum()
```

```python
null_perc = round(100*(df.isnull().sum()/len(df.index)), 2)
null_perc[ null_perc > 0 ]
```

Dropping of columns with more than 30 % null values in them 

```python
df.drop(null_perc[ null_perc > 30 ].index, axis=1, inplace=True)
```

Identifying Columns which have only single unique values or have all unique values. 

```python
col_unique = df.nunique().sort_values(ascending=False)
print(col_unique)
```

```python
col_list_df = col_unique[col_unique.values == 1]
col_list_df.index
```

```python
df.drop(columns = list(col_list_df.index), inplace=True)
df.shape
```

```python
df.loan_status.unique()
```

```python
#Since loan status "Current" doesnt give any info for our analysis for approving or rejecting application,hence filtering out records with status = Current

df = df[df.loan_status != "Current"]
```

```python
# Also id,member_id,url have all unique values we will drop them as they wont help us .
#emp_title, and title also has constant unique values
# total_rec_late_fee, recoveries, collection_recovery_fee, total_rec_int, delinq_2yrs,total_rec_prncp,last_credit_pull_d, out_prncp_inv,last_pymnt_d, out_prncp variables are generated for approved loan. 
#As we are only interested only in loan application details these columns can be removed. 
#recoveries, collection_recovery_fee columns are only valid for charged off loans. Will removed these colomns.

df.drop(['total_rec_int', 'id','total_rec_prncp', 'member_id', 'url','out_prncp', 'out_prncp_inv','total_rec_late_fee', 'last_credit_pull_d', 'recoveries', 'collection_recovery_fee', 'emp_title', 'title',
         'last_pymnt_d','zip_code','delinq_2yrs','earliest_cr_line','inq_last_6mths'], axis=1, inplace=True)
#mths_since_last_delinq
```

We have dropped all the data columns which are deemed surplus for our analysis

**Data Preparation**

```python
df.info()
```

```python
#Identifying Columns based on data type
integer_columns = df.select_dtypes(include=['int64']).columns
float_columns = df.select_dtypes(include=['float64']).columns
object_columns = df.select_dtypes(include=['object']).columns
#df.columns = df.columns.str.strip()
#df[object_columns] = df.apply(lambda x: x.str.strip())
print("object_columns")
print(object_columns)

print("float_columns")
print(float_columns)

print("integer_columns")
print(integer_columns)
```

```python
# Removing Trailing spaces from string columns
df_str = df[object_columns]
df[object_columns] = df_str.apply(lambda x: x.str.strip())

```

```python
# loan_amnt , funded_amnt , int_rate & revol_util should be float  as they are amount & rate of interest respectively
df[['int_rate','revol_util']].head()
df['int_rate'] = df.int_rate.str.strip('%').astype(float)
df['revol_util'] = df.revol_util.str.strip('%').astype(float)
df['loan_amnt']=df.loan_amnt.astype(float)
df['funded_amnt']=df.funded_amnt.astype(float)
```

```python
# Creating a new column to identify the loan status numerically 
df['loan_status_num'] = df['loan_status'].apply(lambda x: 1 if x=='Charged Off' else 0)
```

```python
#issue_d has date values, we need to convert from string  type to date.
df['issue_d'] = pd.to_datetime(df.issue_d, format='%b-%y')
#any data before unix timestamp will be converted incorrectly henceforth subtracting values more than 2020 with 100.
df['issue_d'] = df['issue_d'].apply(lambda x: x-pd.DateOffset(years=100) if x.year > 2020 else x)
df['issue_d_year'] = df.issue_d.dt.year
df['issue_d_month'] = df.issue_d.dt.strftime('%b')
```

```python
#Checking for missinging values and probable imputing them
null_perc = round(100*(df.isnull().sum()/len(df.index)), 2)
null_perc[ null_perc > 0 ]
```

```python
print(df.emp_length.mode())
print(df.revol_util.mode())
print(df.pub_rec_bankruptcies.mode())
```

```python
#since the % of missing values are less , lets impute them with mode value
df.emp_length.fillna(df.emp_length.mode()[0], inplace = True)
df.dropna(subset=['revol_util', 'pub_rec_bankruptcies'], inplace=True)
```

```python
#Identifying Columns based on data type
integer_columns = df.select_dtypes(include=['int64']).columns
float_columns = df.select_dtypes(include=['float64']).columns
object_columns = df.select_dtypes(include=['object']).columns

print("object_columns")
print(object_columns)

print("float_columns")
print(float_columns)

print("integer_columns")
print(integer_columns)
```

```python
# A generic Funtion to calculate the loan default ration
def ratio_cal(col_name):
  df_tmp = df.groupby([col_name, 'loan_status']).loan_status.count().unstack().fillna(0).reset_index()
  df_tmp['Total'] = df_tmp['Charged Off'] + df_tmp['Fully Paid'] 
  df_tmp['ratio'] = df_tmp['Charged Off'] / df_tmp['Total']
  df_tmp.sort_values('ratio', ascending=False)
  return df_tmp
```

```python
#Creating a new range col for loan amount based on percentile distribution of loan amount
df.loc[df['loan_amnt'].between(500, 5000, 'both'), 'loan_amnt_range'] = 'low' 
df.loc[df['loan_amnt'].between(5000, 11000, 'right'), 'loan_amnt_range'] = 'medium'
df.loc[df['loan_amnt'].between(11000, 16000, 'right'), 'loan_amnt_range'] = 'high'
df.loc[df['loan_amnt'].between(17000, 35000, 'right'), 'loan_amnt_range'] = 'very high'
plt.figure(figsize=(10,6))
sns.barplot(x="loan_amnt_range",data=ratio_cal('loan_amnt_range'),y='ratio')
plt.title('loan default_ratio vs Loan Amount')
plt.xlabel('Loan Amount')
plt.ylabel('loan default_ratio')      

plt.show()
```

Higher the loan amount, greater the chance of the loan getting default.
